```python
from sklearn.datasets import fetch_olivetti_faces

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# face images 
- of 40 people 
- 10 images per person
- 64*64 pixel per image

```python
# fetch the faces data
faces = fetch_olivetti_faces()

faces.images.shape, faces.data.shape, faces.target.shape
```

```python
plt.imshow(faces.images[1,:, :]);

# or plt.imshow(faces.images[1]);
```

```python
plt.hist(faces.images[0,:,:].ravel());
```

```python
plt.imshow(faces.data[0].reshape(64,64));
```

```python
# basic grid adjustment
fig = plt.figure(figsize=(8, 8))


# fig.subplots_adjust(top=0.95, wspace = 0.35, hspace=0.55)
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(400):
    ax = fig.add_subplot(20, 20, i + 1, xticks=[], yticks=[])
    ax.imshow(faces.images[i], cmap=plt.cm.bone)

# or,
# plt.imshow(faces.images[0,:, :])
```
