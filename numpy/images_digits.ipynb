```python
from sklearn.datasets import load_digits

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# fetch the faces data
digits = load_digits()

digits.images.shape, digits.data.shape, digits.target.shape
```

## display data

```python
# basic grid adjustment
fig = plt.figure(figsize=(6, 6))


# fig.subplots_adjust(top=0.95, wspace = 0.0035, hspace=0.00000055)
fig.subplots_adjust(left=0, right=1, bottom=0, top=0.35, hspace=0.05, wspace=0.05)

for i in range(300):
    ax = fig.add_subplot(10, 30, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary)

# or,
# plt.imshow(faces.images[0,:, :])
```
