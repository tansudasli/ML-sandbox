```python
import cv2

import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
```

## basic coloring and drawing on images

```python
# in RGB format
#  'rand': np.random.randint(0, high=256, size=(3,)).tolist(), 

colors = {'blue': (0, 0, 255), 'green': (0, 255, 0), 'red': (255, 0, 0), 'yellow': (255, 255, 0), 
          'magenta': (255, 0, 255), 'cyan': (0, 255, 255), 'white': (255, 255, 255), 
          'black': (0, 0, 0), 
          'gray': (125, 125, 125), 'dark_gray': (50, 50, 50), 'light_gray': (220, 220, 220)}

colors['blue']
```

```python
# test colors
image = np.zeros((500, 500, 3), dtype="uint8")

separation = 20
# :)
# for key in colors:
#     image[separation:separation+20,:] = colors[key]
#     separation+= 40

for key in colors:
    cv2.line(image, (0, separation), (500, separation), colors[key], 10)  #gets x,y as pt1!
    separation += 40
```

## image

```python
img = cv2.imread('images/lena.jpg', cv2.IMREAD_COLOR)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #opencv reads BGR format default


img.shape
```

```python
cv2.rectangle(img, (200, 200), (380, 400), colors['yellow'], 3)
```

```python
images = [image, img ]
titles = ['colors', 'orginal']
```

```python
# basic grid adjustment
fig = plt.figure(figsize=(12, 12))

fig.subplots_adjust(left=0, right=1, bottom=0, top=0.69, hspace=0.05, wspace=0.25)

for i in range(2):
    ax = fig.add_subplot(1, 4, i + 1) #, xticks=[], yticks=[]
    ax.set_title(titles[i])
    ax.imshow(images[i], cmap=plt.cm.gray)
```
