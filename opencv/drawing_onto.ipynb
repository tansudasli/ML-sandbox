```python
import cv2

import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# in RGB format
#  'rand': np.random.randint(0, high=256, size=(3,)).tolist(), 

colors = {'blue': (0, 0, 255), 'green': (0, 255, 0), 
          'red': (255, 0, 0), 'yellow': (255, 255, 0), 
          'magenta': (255, 0, 255), 'cyan': (0, 255, 255), 
          'white': (255, 255, 255), 'black': (0, 0, 0), 
          'gray': (125, 125, 125), 
          'dark_gray': (50, 50, 50), 'light_gray': (220, 220, 220)}

colors['blue']
```

```python
# test colors
image = np.zeros((500, 500, 3), dtype="uint8")

seperation = 20
for color in colors:
    image[seperation:seperation+20,:] = colors[color]
    seperation+= 40

plt.imshow(image)
```
