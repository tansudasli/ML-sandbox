# gradient_edge_detection

- image gradient (direction change in intensity or color)
  - useful for edge detection
  
##### Methods
- soble
- scharr
- laplacian


```python
import cv2

import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('images/messi5.jpg', cv2.IMREAD_GRAYSCALE)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #opencv reads BGR format default


img[0,0]
```

```python
lap = cv2.Laplacian(img, cv2.CV_64F ) #, ksize=5
lap = np.uint8(np.absolute(lap))

sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)
sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 1)

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
sobel = np.uint8(np.absolute(sobel))

lap[0,0]
```

```python
images = [img, lap, sobelX, sobelY, sobel ]
titles = ['orginal', 'laplacian', 'sobelX', 'sobelY', 'sobel']
```

```python
# basic grid adjustment
fig = plt.figure(figsize=(24, 12))

fig.subplots_adjust(left=0, right=1, bottom=0, top=0.69, hspace=0.05, wspace=0.05)

for i in range(5):
    ax = fig.add_subplot(2, 4, i + 1,) # xticks=[], yticks=[]
    ax.set_title(titles[i])
    ax.imshow(images[i], cmap=plt.cm.gray)
```
