```python
import cv2

import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('images/LinuxLogo.jpg', cv2.IMREAD_COLOR)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #opencv reads BGR format default


img.shape
```

```python
h,w = img.shape[:2]

mask = np.zeros((h+2, w+2), np.uint8)

img_f = img.copy()
x = cv2.floodFill(img_f, mask, (50,50), (255,0,0), (6,6,6), (6,6,6) )
```

```python

images = [img, mask, img_f]
titles = ['orginal', 'mask', 'flood_fill']
```

```python
# basic grid adjustment
fig = plt.figure(figsize=(12, 12))

fig.subplots_adjust(left=0, right=1, bottom=0, top=0.69, hspace=0.05, wspace=0.05)

for i in range(3):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.set_title(titles[i])
    ax.imshow(images[i], cmap=plt.cm.gray)
```
