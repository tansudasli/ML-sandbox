```python
import cv2

import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('images/lena.jpg', cv2.IMREAD_COLOR)

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  #opencv reads BGR format default


img.shape
```

```python

integral = cv2.integral(img)
integral_normalized = (255.0*integral) / integral.max() #normalize
```

```python
images = [img, integral, integral_normalized ]
titles = ['orginal', 'integral', 'integral_normalized']
```

```python
# basic grid adjustment
fig = plt.figure(figsize=(12, 12))

fig.subplots_adjust(left=0, right=1, bottom=0, top=0.69, hspace=0.05, wspace=0.05)

for i in range(3):
    ax = fig.add_subplot(1, 4, i + 1, xticks=[], yticks=[])
    ax.set_title(titles[i])
    ax.imshow(images[i], cmap=plt.cm.gray)
```
