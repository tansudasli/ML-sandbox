```python
import cv2

import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('images/opencv-logo.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #opencv reads BGR format default


img.shape
```

```python
kernel = np.ones((5,5), np.float32)/ 25  # divide by 5*5 = 25

dst = cv2.filter2D(img, -1, kernel)
```

```python
images = [img, dst]
titles = ['orginal', '2D convolution']
```

```python
# basic grid adjustment
fig = plt.figure(figsize=(10, 10))

fig.subplots_adjust(left=0, right=1, bottom=0, top=0.65, hspace=0.05, wspace=0.05)

for i in range(2):
    ax = fig.add_subplot(1, 2, i + 1, xticks=[], yticks=[])
    ax.set_title(titles[i])
    ax.imshow(images[i], cmap=plt.cm.gray)
```
