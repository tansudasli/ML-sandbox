# Smoothing

- blurring

```python
import cv2

import numpy as np 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# img = cv2.imread('images/opencv-logo.png', cv2.IMREAD_COLOR)
img = cv2.imread('images/water.png', cv2.IMREAD_COLOR)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #opencv reads BGR format default


img.shape
```

```python
kernel = np.ones((5,5), np.float32)/ 25  # divide by 5*5 = 25

dst = cv2.filter2D(img, -1, kernel)
blur = cv2.blur(img, (5,5))   #similar to filter2D
gaussian_blur = cv2.GaussianBlur(img, (5,5), 0) # removes noise also
median_blur = cv2.medianBlur(img, 3) #should be odd number. removes salt&pepper noise. also smoothes edges
```

```python
images = [img, dst, blur, gaussian_blur, median_blur]
titles = ['orginal', '2D convolution', 'blur', 'gaussian_blur', 'median_blur']
```

```python
# basic grid adjustment
fig = plt.figure(figsize=(12, 12))

fig.subplots_adjust(left=0, right=1, bottom=0, top=0.69, hspace=0.05, wspace=0.05)

for i in range(5):
    ax = fig.add_subplot(2, 4, i + 1, xticks=[], yticks=[])
    ax.set_title(titles[i])
    ax.imshow(images[i], cmap=plt.cm.gray)
```
