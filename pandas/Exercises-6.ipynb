### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
df1 = sales1.rename(columns={
    'Book title': 'title',
    'Number sold': 'number',
    'Sales price': 'their_price',
    'Royalty paid': 'royalty',
    })
df1['currency'] = 'USD'
df1['source'] = 'other'
df1
```

```python
s = sales2.copy()

t = sales2['Title']
t = t.where(t.str.endswith(')')).str.split().str[-1].str.strip('()')   # get USD


s['currency'] = t.fillna(method='bfill')    #fill before u see USD/GBP ..
s = s[s['List price'].notnull()]
s['source'] = 'aws'

df2 = s.rename(columns={
    'Units sold': 'number',
    'Title': 'title',
    'List price': 'their_price',
    'Royalty': 'royalty',
    })
s
```

```python
df = pd.concat([df1, df2])
df
```
